**4. Calculamos la carrera más cara y más barata**

 **7. Calculamos las relación tiempo/km más alta y más baja**

```python
dg3.dtypes
```

```python
dg3['Time_carreer_seconds'] = dg3.apply(
    lambda s:
    time.mktime((datetime.strptime(s['Lpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S")).timetuple())
    -time.mktime((datetime.strptime(s['lpep_pickup_datetime'],"%Y-%m-%d %H:%M:%S")).timetuple())
    ,axis=1)
```

```python
dg3.head()
```

```python
dg3['Time_carrer_minutes'] = dg3.apply(
    lambda s:
    (s['Time_carreer_seconds'] % 3600) // 60
    ,axis=1)
```

```python
dg3.head()
```

```python
dg3['Time_long_vs_km'] = dg3.apply(
    lambda s: round(s['Time_carreer_seconds']/s['Trip_distance'],2), axis=1)
```

```python
expensive = dg3[dg3['Time_long_vs_km'] ==  np.max(dg3['Time_long_vs_km'])]
```

**1. Creamos variables relevantes para el caso de estudio**

```python
dg3['Week_day'] = dg3.apply(
    lambda s:
    datetime.weekday(datetime.strptime(s['Lpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S"))+1
    ,axis=1)
```

```python
dg3.head()
```

```python
dg3['Hour_day'] = dg3.apply(
    lambda s:
    datetime.strptime(s['Lpep_dropoff_datetime'],"%Y-%m-%d %H:%M:%S").hour
    ,axis=1)
```

```python
dg3.head()
```

```python
def df(d,dia,hora):

    return d[(d['Week_day']==dia) & (d['Hour_day']==hora)]
```

```python
import matplotlib.pyplot as plt
from pylab import *
```
